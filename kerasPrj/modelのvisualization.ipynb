# modelのvisualization

## VGG16
### PNGのイメージ

```python
from keras.utils import plot_model
from tensorflow.python.keras.utils.vis_utils import plot_model
from keras.applications.inception_v3 import InceptionV3
from keras.applications import vgg16

vgg = vgg16.VGG16(include_top=True, weights='imagenet')
layers = [(layer, layer.name, layer.trainable) for layer in vgg.layers]
plot_model(vgg, to_file='model_vgg16.png',show_shapes=True)
```

```python
from IPython.display import Image
Image('model_vgg16.png')
```

### layer名の表示

```python
for l, layer in enumerate(layers):
    print(l, " ", layer)
```

## InceptionV3

```python
icpv3 = InceptionV3(weights='imagenet', include_top=True)
plot_model(icpv3, to_file='model_InceptionV3.png',show_shapes=True)
Image('model_InceptionV3.png')
```

## ResNet50

```python
from keras.applications import resnet50

resnet50 = resnet50.ResNet50(weights='imagenet', include_top=True)
plot_model(resnet50, to_file='model_resnet50.png',show_shapes=True)
Image('model_resnet50.png')
```
